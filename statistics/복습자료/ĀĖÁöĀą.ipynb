```python
pcl_grp = df.groupby('gender') #성별로 그룹화
```

```python
pcl_grp.get_group('female')
```

```python
pcl_grp
```

```python
for i, j in pcl_grp:
  print(i, '묶인 그룹')
  print(j, 'data')
```

```python
pcl_grp_2 = df.groupby(['part_time_job', 'absence_days'])
for i, j in pcl_grp_2:
  print(i, '묶인 그룹')
  print(j, 'data')
```

```python
pcl_grp_2.get_group((False, 0)) #알바를 하지 않고 결석한 적이 없는 사람들
```

```python
#결석 일수에 따른 물리 점수 평균 추출하기
df.groupby('absence_days')[['physics_score']].mean()
```

```python
#결석일수와 성별에 따른 물리 점수 평균 추출하기
df.groupby(['absence_days', 'gender'])[['physics_score']].mean()
```

```python
df.groupby(['absence_days', 'gender'])[['physics_score']].agg(['min', 'max', 'std'])
#결석일수와 성별에 따른 물리점수의 최소, 최대, 표준편차 보기
```

```python
df['gender'] == 'female' #여성인가
```

```python
df[df['gender'] == 'male'] #남성 추출
```

```python
#여성이고 3일 결석한 사람들의 주간자율학습시간 평균은?
df[(df['gender']=='female') & (df['absence_days']==3)]['weekly_self_study_hours'].mean()
```

```python
df[(df['gender']=='male') & (df['part_time_job']==True)].groupby('absence_days')['extracurricular_activities'].sum()
#남성이고 알바를 하고 있는 사람들을 결석일수에 따라 그룹화 후 과외 활동하는 경우의 합
```

```python
#알바를 하지 않고 생물 점수가 89점인 경우 추출하기
df[(df['part_time_job'] == False) & (df['biology_score'] == 89)]
```
