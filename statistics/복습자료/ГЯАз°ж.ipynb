## agg 함수
- 여러가지 통계치를 함께 보기 위해 사용 (요약통계치)
      group.agg(['통계1','통계2','통계3'])
  - count: 각 그룹의 항목 개수

```python
pcl_grp2.agg(['min','max','count'])
```

```python
car_grp2.agg(['min','max','count'])
```

```python
# agg함수를 이용해서 내가 직접 함수를 만들거나 계산식을 만들어서 적용시킬 수 있다.

def min_max(x): # range값 계산 함수가 됨 - 여기서는 별 의미없는 값
    return x.max()- x.min()
```

```python
pcl_grp.agg(min_max) # 내가 만든 계산식을 agg함수를 이용해서 적용시킬 수 있다.
```

```python
car_grp.agg(min_max)
```

# 데이터 병합
- 판다스에서 제공하는 merge, concat 함수
- merge 기준 컬럼을 통한 ( 공통된 컬럼( 값 기준 )) 데이터를 병합
    - 공통된 **기준 값인 컬럼**이 필요하다
    - join 할 수 있는 방법이 다양
      - left, right, outer, inner
        - inner: 두 데이터프레임 간에 일치하는 키(열)를 기준으로 조인을 수행하고, 결과로서 두 데이터프레임에서 모두 일치하는 행만을 유지
          - 교집합
        - outer 조인: 두 데이터프레임 중 하나에 해당하는 모든 행을 포함; 일치하는 경우에는 일치하는 값으로, 일치하지 않는 경우에는 NaN(결측치)으로 채워짐
          - 합집합
        - right: 오른쪽 데이터프레임을 기준으로 조인; 오른쪽 데이터프레임의 모든 행을 유지하고, 왼쪽 데이터프레임과 일치하는 경우에만 결합; 왼쪽 데이터프레임에 해당하는 데이터가 없는 경우에는 NaN(결측치)으로 채워짐
        - left: 왼쪽 ...
    - 공통 컬럼 기준으로 여러 컬럼을 합칠 수 있다.
- concat 데이터와 데이터 덩어리와 덩어리를 합쳐서 병합한다.
    - df1, df2를 예시로 붙인다면? 행과 열 기준으로 붙일 수 있다.
      - axis =0, 1에 따라서 붙이는 방법이 다르다.
        - df1 + df2 ....+ dfn (가로로 붙이는 경우)
        - 아래로... (세로로 붙이는 경우)
    
- 데이터 병합을 하는 이유는?
  - 고객이 어떤 주문을하고 어떤 행사에 참여했는지를 보고 싶다 -> 고객+주문+행사를 합쳐서 봐야 한다.

```python
test1=pd.DataFrame({'학회원번호':[1,2,3,4,5],
             '학회원분반':['Python','SQL','R','데이터분석입문','데이터분석전처리']})
```

```python
test2=pd.DataFrame({'학회원번호':[1,2,3,4,5],
             '상점현황':[4,3,5,10,2]})
```

## pd.merge()
- join function
- 함수는 두 개의 데이터프레임을 하나로 결합할 때 특정 열(또는 인덱스)을 기준으로 조인하는 데 사용
  - 데이터프레임 간에 공통된 열(또는 인덱스)을 필요로 함

```python
pd.merge(test1, test2, how='inner')
```

## pd.concat()
- 연결 function
- 하나 이상의 데이터프레임을 단순히 연결하여 새로운 축으로 합칠 때 사용되며, 행 또는 열의 순서를 바꿀 수 있음

```python
# concat
pd.concat([test1,test2]) # 디폴트는 행기준으로 붙여진다. (axis = 0: 아래로)
```

```python
pd.concat([test1,test2],axis=1) # 옆으로
```
