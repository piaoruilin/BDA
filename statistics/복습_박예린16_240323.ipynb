```python
import pandas as pd
import seaborn as sns

titanic = sns.load_dataset('titanic')
df=titanic.copy()

df
```

```python
pcl_grp=df.groupby('pclass') #요약통계치
pcl_grp
```

```python
for i, j in pcl_grp:
    print(i, '묶인 그룹')
    print(j, 'data')
```

```python
pcl_grp.get_group(2)
```

```python
pcl_grp2=df.groupby(['pclass','sex'])
for i,j in pcl_grp2:
    print(i, '묶인 그룹')
    print(j, 'data')
```

```python
pcl_grp2.get_group((3, 'female'))
```

```python
pcl_grp2.mean()
pcl_grp.std()
pcl_grp2.agg(['min','max','count'])
```

```python
def min_max(x):
    return x.max() - x.min()

pcl_grp.agg(min_max)
```
