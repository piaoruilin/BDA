```python
import seaborn as sns
tt=sns.load_dataset('titanic')
```

```python
tt.query("""embarked=='S'""") #1st condition
```

```python
tt.query("""pclass == 1 """) #2nd condition
```

```python
tt.query("""sex == 'female' """) #3rd condition
```

```python
tt.query(""" age < 20 """)
```

```python
#Q1
tt.query("""embarked == 'S' and  pclass == 1  and sex == 'female' and  age < 20 """)['fare'].mean()
```

```python
tt.query("""embarked == 'S' and  pclass == 1  or sex == 'female' or  age < 20 """)['fare'].mean()
```

```python
tt.query("""embarked == 'S' and  pclass == 1  or sex == 'female' or  age < 20 """)[['alive','fare','age']].loc[::2].fare.mean()
```
